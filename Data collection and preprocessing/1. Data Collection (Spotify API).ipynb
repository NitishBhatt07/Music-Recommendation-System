### Authentication of Spotify API

```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="3e14477f91304445a23659600f1c43cc",
                                                           client_secret="d3fd764442b242d98086f5627a6c6c90"))
```

### Get All Tracks

```python
def get_playlist_tracks(username,playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

### Get Info DataFrame

```python
def getPlaylistInfoDf(tracks):
    
    track_all_info = []
    track_main_info = []
    
    temp_df = pd.DataFrame(columns=["track_main_info","track_all_info"])

    for track in tracks:
        # Main Artist
        artist_uri = track["track"]["artists"][0]["uri"]
        artist_info = sp.artist(artist_uri)
        
        track_main_info.append(artist_info)
        track_all_info.append(track)

    temp_df["track_all_info"] = track_all_info
    temp_df["track_main_info"] = track_main_info
    
    
    return temp_df
```

### Get Main DataFrame with all track info

```python
def getMainDataFrame(temp_df):
    new_df = pd.DataFrame(columns=["id","song","album","singers","genres","popularity","decade","thumbnail"])
    
    song_id = []
    song_name = []
    song_album = []
    singer_name = []
    song_popularity = []
    song_genre = []
    song_thumbnail = []
    song_decade = []
    
    # Get Song Name , Album Name
    for data in temp_df["track_all_info"]:
        song = data['track']['name'].split("(")[0].split("-")[0]
        album = data['track']['album']['name'].split("(")[0].split("-")[0]
        decade = data["track"]['album']['release_date'][:2]
        song_name.append(song)
        song_album.append(album)
        song_decade.append(decade)
    
    # Get Singer Name
    for i in range(len(temp_df['track_all_info'])):
        all_singer = []
        for j in range(len(temp_df['track_all_info'][i]['track']['artists'])):
            all_singer.append(temp_df['track_all_info'][i]['track']['artists'][j]['name'])
        singer_name.append(all_singer) 
    
    # Get Popularity, Genres and Thumbnail
    for data in temp_df['track_main_info']:
        songId = data["id"]
        genres = data['genres']
        popularity = data['popularity']
        image = data['images'][0]['url']
        song_id.append(songId)
        song_genre.append(genres)
        song_popularity.append(popularity)
        song_thumbnail.append(image)
    
    new_df["id"] = song_id
    new_df["song"] = song_name
    new_df["album"] = song_album
    new_df["singers"] = singer_name
    new_df["genres"] = song_genre
    new_df["popularity"] = song_popularity
    new_df["decade"] = song_decade
    new_df["thumbnail"] = song_thumbnail
    
    return new_df    
```

### Program Flow 

```python
def allPlaylistLink(filename):
    with open(filename) as f:
        all_playlist_link = f.readlines()
    return all_playlist_link 
```

```python
filename = "Playlists.txt"
all_playlist_link = allPlaylistLink(filename)
```

```python
main_df = pd.DataFrame(columns=["id","song","album","singers","genres","popularity","decade","thumbnail"])

for playlist_link in all_playlist_link:
    
    tracks = get_playlist_tracks("3e14477f91304445a23659600f1c43cc",playlist_link)
    temp_df = getPlaylistInfoDf(tracks)
    new_df = getMainDataFrame(temp_df)
    main_df = pd.concat([main_df,new_df])
```
